#### Relaciones 



##### Calidad Sueño y Profesión


Se puede observar que las carreras relacionadas a los negocios tienden a tener una peor calidad del sueño mientras que las carreras dei ingenieria son las que mejor calidad del sueño presentan.



##### Sueño y Actividad Física 

Podemos observar que la duraicón del sueño aumenta en función al nivel de activiadad física y la calidad del sueño alcanza el nivel más alto cuando la actividad física es mayor.




#### Profesión con mayor nivel de estres y Indice de Masa Corporal 

La profesión con mayor nivel de estres y IMC es la de representante de ventas y la de menor IMC y nivel de estres es la de ingenieria, siendo también la que mayor calidad del sueño posee. 






#### Cantidad de Trasntornos del Sueño 

La frecuencia de transtornos del sueño va aumentando desde los 18 años siendo su pico entre los 40-50 años con 93 casos para luego ir descendiendo.



#### Cantidad de Pasos al Día 

A medida que aumenta la cantidad del sueño y el IMC lo hacen de manera indirecta. Es decir, mientras más pasos al día de una persona este tiende a tener una mejor calidad del sueño y un menor IMC  y a menor cantidad de pasos al día es mayor el IMC y menor la calidad del sueño. 



#### Conclusión

- Entre los 30 y 70 años las personas es donde más transtornos del sueño experimentan, siendo el pico entre los 40 y 50 años.

- Entonces el nivel de activiad física influye positivamente tanto en la duración como en la calidad del sueño reduciendo los niveles de estres y el IMC en las personas , siendo así que las profesiones con peor calidad del sueño son las que presentan mayores niveles de estres y mayo IMC.

- Entonces si una persona una persona posee altos niveles de estres y un alto IMC, muy probablemente esta persona este durmiendo mal y tenga una baja actividad fisica. 

- Concluyendo que si se quisiera mejorar el sueño y reducir el estres, se tendría que aumentar el nivel de actividad física en el paciente. Ahora esto no es totalmente determinante ya que los datos poseen una variabilidad sifnificativa que podrian indicarnos que otros facotres influyen en esta relación, pero, aún así la tendencia general sigue obedece a este patron.
